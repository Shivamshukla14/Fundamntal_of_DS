```python
import matplotlib.pyplot as plt

plt.figure(figsize=[15,5])

countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']

totalCases = (2026493, 710887, 476658, 5971, 27136, 40597, 7473)

for i in range(len(countries)):
    plt.bar(countries[i], totalDeaths[i], bottom = totalCases[i] - totalDeaths[i], color = 'black')
    plt.bar(countries[i], totalCases[i] - totalDeaths[i], color = 'red')

plt.legend(labels = ['Total Deaths', 'Total Cases'])

plt.title("Bar plot representing the total deaths")

plt.xlabel('Countries')
plt.ylabel('Cases')

plt.savefig('3BarPlot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[15,10])

countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']
totalDeath=[113055,37312,5971,7473,33964,345345]
activeCases = [1139958,347973,239999,129360,34730,55686]
totalRecovery=[773480,325602,230688,129095,166584,456456]

x= np.arange(len(totalDeath))
plt.bar(x,totalDeath,color='r',width=0.25)
plt.bar(x+0.25,totalRecovery,color='g',width=0.25)
plt.bar(x+0.5,activeCases,color='b',width=0.25)
plt.legend(['Total Deaths','Total Recovery','Active Cases'])
plt.xticks([i+0.25 for i in range(6)],countries)
plt.title("Bar plot representing the total Deaths, total recovered cases")
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('4barplot.png')
plt.show()
```
